```python
df['Maximum Open Credit'].isnull().sum()
```

```python
df['Current Credit Balance']=pd.cut(x = df['Current Credit Balance'],
                        bins = [-1,0,250000,1000000,2000000,2500000,6000000,12000000], 
                        labels = ["Credit Balance: None","Credit Balance: Normal","Credit Balance: Medium", "Credit Balance: High", 
                                  "Credit Balance: Very High", "Credit Balance: Extreme", "Credit Balance: Insane"])
df
```

```python
df['Credit Score']=pd.cut(x = df['Credit Score'],
                        bins = [-1,300,579,669,739,799,850], 
                        labels = ["Credit Score: Very Poor","Credit Score: Poor","Credit Score: Fair","Credit Score: Good", "Credit Score: Very Good","Credit Score: Exceptional"])
df
```

```python
df['Monthly Debt']=pd.cut(x = df['Monthly Debt'],
                        bins = [-1,0,25000,50000,75000,100000,150000,300000], 
                        labels = ["Monthly Debt: None","Monthly Debt: Low","Monthly Debt: Normal","Monthly Debt: Medium", "Monthly Debt: High","Monthly Debt: Very High","Monthly Debt: Extreme"])
df
```

```python
df['Annual Income']=pd.cut(x = df['Annual Income'],
                        bins = [-1,0,2000000,5000000,10000000,35000000], 
                        labels = ["Annual Income: None","Annual Income: Normal","Annual Income: Medium", "Annual Income: High", "Annual Income: Very High"])
df
```

```python
df['Current Loan Amount']=pd.cut(x = df['Current Loan Amount'],
                        bins = [-1,0,10000,100000,500000,1000000,100000000], 
                        labels = ["Current Loan Amount: None","Current Loan Amount: Low","Current Loan Amount: Medium", "Current Loan Amount: High", "Current Loan Amount: Very High","Current Loan Amount: Extreme"])
df
```

```python
df['Years in current job'] = df['Years in current job'].replace({'< 1 year': 'Current Job: Less Than One Year', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'1 year': 'Current Job: One Year', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'2 years': 'Current Job: 2-5 Years', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'3 years': 'Current Job: 2-5 Years', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'4 years': 'Current Job: 2-5 Years', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'5 years': 'Current Job: 2-5 Years', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'6 years': 'Current Job: 5-10 Years', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'7 years': 'Current Job: 5-10 Years', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'8 years': 'Current Job: 5-10 Years', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'9 years': 'Current Job: 5-10 Years', True: False})  # raises
df['Years in current job'] = df['Years in current job'].replace({'10+ years': 'Current Job: More Than Ten Years', True: False})  # raises
df
```

```python
df['Years of Credit History'] = df['Years of Credit History'].apply(np.int64)
df['Years of Credit History'].unique()

df['Years of Credit History']=pd.cut(x = df['Years of Credit History'],
                        bins = [-1,10,20,50,100], 
                        labels = ["Credit History: Up to Ten","Credit History: Up to Twenty", "Credit History: Up to Fifty", "Credit History: More than Fifty"])
df
```

```python
df['Number of Open Accounts'] = df['Number of Open Accounts'].apply(np.int64)
df['Number of Open Accounts'].unique()

df['Number of Open Accounts']=pd.cut(x = df['Number of Open Accounts'],
                        bins = [-1,10,20,50], 
                        labels = ["Open Accounts: Up to Ten","Open Accounts: Up to Twenty", "Open Accounts: More than Twenty"])
df
```

```python
df['Number of Credit Problems'] = df['Number of Credit Problems'].apply(np.int64)
df['Number of Credit Problems'].unique()

df['Number of Credit Problems']=pd.cut(x = df['Number of Credit Problems'],
                        bins = [-1,0,3,15], 
                        labels = ["Credit Problems: None","Credit Problems: Few", "Credit Problems: Many"])
df
```

```python
df['Tax Liens'] = df['Tax Liens'].apply(np.int64)
df['Tax Liens'].unique()

df['Tax Liens']=pd.cut(x = df['Tax Liens'],
                        bins = [-1,0,10,15], 
                        labels = ["Tax Liens: None","Tax Liens: UpToTen", "Tax Liens: More Than Ten"])
df
```

```python
df['Bankruptcies'].unique()

df['Bankruptcies']=pd.cut(x = df['Bankruptcies'],
                        bins = [-1,0,7], 
                        labels = ["Bankruptcies: None","Bankruptcies: At Least One"])
df
```
